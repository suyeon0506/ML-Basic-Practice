```python
import tensorflow as tf

X_t = [1,2,3]
Y_t = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name = 'weight') #보통 W는 대문자
b = tf.Variable(tf.random_normal([1]), name = 'bias')
hypothesis = W * X_t + b
cost = tf.reduce_mean(tf.square(hypothesis - Y_t)) #Y값 빼주는거 잊지말기

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.1)
train = optimizer.minimize(cost) #여기는 tf. 필요 없음

sess = tf.Session()
sess.run(tf.global_variables_initializer()) #초기화 꼭 필요, initializer임

for step in range(2001):
    sess.run(train)
    if(step % 20 == 0):
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```
